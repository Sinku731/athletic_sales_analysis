### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")

# Display the first few rows of each DataFrame
print(sales_2020_df.head())

print("\n2021 Data:")
print(sales_2021_df.head())
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df], ignore_index=True)
#Reset the index of the combined DataFrame
combined_sales_df.reset_index(drop=True, inplace=True)
# Display the first few rows of the combined DataFrame
combined_sales_df.head()
```

```python
# Check if any values are null.
null_values = combined_sales_df.isnull().sum()
print(null_values)
```

```python
# Check the data type of each column
data_types = combined_sales_df.dtypes
print(data_types)
```

```python
# Convert the "invoice_date" to a datetime datatype (Table)
combined_sales_df["invoice_date"] = pd.to_datetime(combined_sales_df["invoice_date"])
print(combined_sales_df.head())
print(combined_sales_df.dtypes)
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_sales_df.info())
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
grouped_sales = combined_sales_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
grouped_sales.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
grouped_sales = grouped_sales.sort_values(by='Total_Products_Sold', ascending=False)


# Show the top 5 results.
print(grouped_sales.head())
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold_by_location = combined_sales_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()


# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_by_location.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

top_5_products_sold = products_sold_by_location.sort_values(by='Total_Products_Sold', ascending=False).head(5)

# Show the top 5 results.
print(top_5_products_sold)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_by_location = combined_sales_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
sales_by_location = sales_by_location.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
top_5_sales = sales_by_location.sort_values(by='Total Sales', ascending=False).head(5)
print(top_5_sales)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_by_location = combined_sales_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()


# Optional: Rename the "total_sales" column to "Total Sales"
sales_by_location = sales_by_location.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
top_5_sales = sales_by_location.sort_values(by='Total Sales', ascending=False).head(5)
print(top_5_sales)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
sales_by_retailer_location = combined_sales_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
sales_by_retailer_location = sales_by_retailer_location.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
top_5_sales_retailer_location = sales_by_retailer_location.sort_values(by='Total Sales', ascending=False).head(5)
print(top_5_sales_retailer_location)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_sales = combined_sales_df.pivot_table(
    values='total_sales', 
    index=['retailer', 'region', 'state', 'city'], 
    aggfunc='sum'
).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_sales = pivot_sales.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
top_5_pivot_sales = pivot_sales.sort_values(by='Total Sales', ascending=False).head(5)
print(top_5_pivot_sales)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data for Women's Athletic Footwear
women_athletic_footwear_sales = combined_sales_df[combined_sales_df['product'] == "Women's Athletic Footwear"]

# Select and display only the first 5 rows for simplicity
women_athletic_footwear_sales_summary = women_athletic_footwear_sales[['retailer', 'retailer_id', 'invoice_date', 'region', 'state', 'city', 'product', 'price_per_unit', 'units_sold', 'total_sales', 'operating_profit', 'sales_method']]

# Show top 5 results (you can adjust this if needed)
print(women_athletic_footwear_sales_summary.head())
```

#### Using `groupby`

```python
 #Group by retailer, region, state, and city, then sum the 'units_sold'
womens_footwear_sales = combined_sales_df[combined_sales_df["product"] == "Women's Athletic Footwear"] \
    .groupby(["retailer", "region", "state", "city"])["units_sold"] \
    .sum() \
    .reset_index()

# Rename the 'units_sold' column to 'Womens_Footwear_Units_Sold'
womens_footwear_sales.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)

# Show the top 5 results
top_5_womens_footwear_sales = womens_footwear_sales.nlargest(5, "Womens_Footwear_Units_Sold")

# Display the result
print(top_5_womens_footwear_sales)

```

#### Using `pivot_table`

```python
# Step 1: Filter for only "Women's Athletic Footwear"
women_athletic_footwear_sales = combined_sales_df[combined_sales_df['product'] == "Women's Athletic Footwear"]

# Step 2: Create the pivot table, summing the 'units_sold' for each combination of retailer, region, state, and city
womens_athletic_footwear_sales_pivot = pd.pivot_table(
    women_athletic_footwear_sales,  # Use the filtered data
    values='units_sold',            # Sum units sold
    index=['retailer', 'region', 'state', 'city'],  # Group by these columns
    aggfunc='sum'                   # Aggregating function: sum
)

# Step 3: Rename the 'units_sold' column to 'Womens_Footwear_Units_Sold'
womens_athletic_footwear_sales_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)

# Step 4: Display the pivot table to check the results before filtering top 5
print(womens_athletic_footwear_sales_pivot.head())  # This prints the first 5 rows of the pivot table

# Step 5: Extract the top 5 results based on 'Womens_Footwear_Units_Sold'
top_5_womens_footwear_sales_pivot = womens_athletic_footwear_sales_pivot.nlargest(5, 'Womens_Footwear_Units_Sold')

# Step 6: Show the top 5 results
top_5_womens_footwear_sales_pivot



```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Ensure the 'invoice_date' column is in datetime format
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'], errors='coerce')

# Extract only the date portion (ignoring the time part) if it's there
combined_sales_df['invoice_date'] = combined_sales_df['invoice_date'].dt.date

# Create a pivot table with 'invoice_date' as the index and 'total_sales' as the values
total_sales_pivot = pd.pivot_table(
    combined_sales_df,  # Use the combined sales data
    values='total_sales',  # Sum of total_sales
    index='invoice_date',  # Index by 'invoice_date'
    aggfunc='sum'  # Aggregating function: sum
)

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Sort the pivot table by index (to ensure dates are in chronological order)
total_sales_pivot.sort_index(inplace=True)

# Show the first few rows of the pivot table
total_sales_pivot.head()  # Show the first 5 rows to verify

```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Filter for only 2021 data
combined_sales_2021 = combined_sales_df[
    (combined_sales_df['invoice_date'].dt.year == 2021)
]

# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = combined_sales_2021.pivot_table(
    values='total_sales',
    index='invoice_date',
    aggfunc='sum'
)
weekly_sales_resampled = weekly_sales.resample('W').sum()

# Sort by total sales descending
weekly_sales_sorted = weekly_sales_resampled.sort_values(by='total_sales', ascending=False)

 #Show top 10 results
print(weekly_sales_sorted.head(10))
```
